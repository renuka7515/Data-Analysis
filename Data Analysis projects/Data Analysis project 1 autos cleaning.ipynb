We'll now change the name of the columns. The column names use camelcase instead of Python's preferred snakecase. For that we will use the DataFrame.columns attribute.

```python
data.columns                                                                                                                  
```

```python
data.columns = ['date_crawled', 'name', 'seller', 'offer_type', 'dollar_price', 'ab_test',
       'vehicle_type', 'registration_year', 'gearbox', 'power_ps', 'model',
       'kilometer', 'registration_month', 'fuel_type', 'brand',
       'unrepaired_damage', 'ad_date_created', 'num_photos', 'postal_code',
       'last_seen']
```

```python
data.head()
```

```python
data.info()
```

Convert dates to date type

```python
data['date_crawled'] = pd.to_datetime(data['date_crawled']).dt.date
data['ad_date_created'] = pd.to_datetime(data['ad_date_created']).dt.date
data['last_seen'] = pd.to_datetime(data['last_seen']).dt.date
```

```python
data
```

```python
data.isnull().sum()
```

we can see that there are 5 columns where they have null values nothing but nan.

As all the values in "num_photos" column is zero and is not related to our analysis .so we can drop the column

```python
print(data["num_photos"].unique())
data.drop("num_photos",axis=1,inplace=True)
```

Filling nan values with mode and constant value

```python
data.fillna({"vehicle_type":"unknown","fuel_type":"not mentioned","gearbox":data["gearbox"].mode()[0],"unrepaired_damage":"not mentioned","model":data["model"].mode()[0]},inplace=True)
```
